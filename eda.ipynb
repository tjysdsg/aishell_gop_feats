```python
import os
import numpy as np
from common import *
from matplotlib import pyplot as plt
```

```python
data = {i: [] for i in range(201 + 1)}
with os.scandir('feats') as it:
        for entry in it:
            if entry.is_file():
                filename = entry.name
                pid = int(filename.split('.')[0])
                data[pid] = np.loadtxt(os.path.join('feats', filename))
```

# Thresholds

```python
initials_pid = [phone2id[i] for i in INITIALS]

for pid, scores in data.items():
    scores = np.asarray(scores)
    if len(scores.shape) == 0 or scores.shape[0] == 0:
        continue
    q = 0.75
    if pid in initials_pid:
        q = 0.25
    print(f'{pid}: {scores.shape[0]} samples,', np.quantile(scores, q))
```
